```python
import matplotlib.pyplot as plt
import numpy as np
from random import uniform
```

# Упражнение 1: Корреляция 

```python
def mnk(x,y):
    n = len(x)
    sum_x = sum(x)
    sum_y = sum(y)
    sum_xy = sum(x*y)
    sum_x2 = sum(x*x)
    sum_y2 = sum(y*y)
    

        
    
    Mx = sum_x/n
    My = sum_y/n
    Mxy = sum_xy/n
    sigmax = (sum_x2/n - Mx**2)**0.5
    sigmay = (sum_y2/n - My**2)**0.5
    
    r = (Mxy - Mx*My)/(sigmax*sigmay)
    
    t = (r*(n-2)**0.5)/((1-r**2)**0.5)
    print("Коэффициент корреляции: %s"%r)
    print("t: %s"%t)
    
    if((n*sum_x2 - sum_x**2) != 0):    
        a = (n*sum_xy - sum_x*sum_y)/(n*sum_x2 - sum_x**2)
    else:
        a = 0
    b = (sum_y - a*sum_x)/n
    return a,b
```

```python
def pic(x,y):
    a, b = mnk(x,y)
    print('a = %s, b = %s'%(a,b))
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 8
    fig_size[1] = 8
    plt.title('')
    plt.ylabel('')
    plt.xlabel('')
    #plt.rcParams.update({'font.size': 12})  #размер шрифта
    plt.grid()
    plt.plot([min(x), max(x)], [a*min(x)+b,a*max(x)+b], color='red')
    plt.scatter(x, y, color='black',s=0.5 )
```

## 1) x, y - две псевдослучайные велечины

```python
n = 10000
x = np.array([uniform(0,10) for i in range(n)])
y = np.array([uniform(0,10) for i in range(n)])
pic(x,y)
```

## 2) x - псевдослучайная величина, y = -0.5x+0.6

```python
n = 10000
x = np.array([uniform(0,10) for i in range(n)])
y = np.array([-0.5*i +0.6 for i in x])
pic(x,y)
```

# 3) x,z - псевдослучайные величины, y = z*x

```python
n = 10000
x = np.array([uniform(0,10) for i in range(n)])
z = np.array([uniform(0,10) for i in range(n)])
y = np.array(x*z)
pic(x,y)
print(" alpha = 0.05 \n t_{0.5} = 1.96 \n тк t<t_{0.5}, то отклоням гипотизу о независимости x и у")
print(" ")
```
