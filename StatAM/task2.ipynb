```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

# Упражнение 2 Центральная предельная теорема

## Гамма- распределение 
## $\mu = k\theta  $
## $\sigma^{2} = k\theta^{2}  $

```python
k = 2.0
theta = 1.0 
n = 10000
sample = np.array([random.gammavariate(k,theta) for i in range(n)])
```

```python
plt.hist(sample, density=True, histtype='stepfilled', alpha=400.1,label='histogram', bins=50)
plt.legend(loc='best', frameon=False)
plt.legend()
```

```python
def get_plot(k, theta,n):
    #k и theta параметры гамма распределения, n размер выборки
    mu = k*theta
    #согласно цпд
    sigma = ((k*theta**2)/n)**0.5
    print("Размер выборки: %s, стандарное отклонение: %s"%(n,sigma))
    matrix = np.empty(10000)
    for i in np.arange(len(matrix)):
        sample = np.array([random.gammavariate(k,theta) for i in range(n)])
        matrix[i]= np.mean(sample)
    
    fig_size = plt.rcParams["figure.figsize"]
    fig_size[0] = 18
    fig_size[1] = 8
    plt.hist(matrix, density=True, histtype='stepfilled', alpha=400.1, label='Размер выборки : %s'%n , bins=60)
    plt.legend(loc='best', frameon=False)
    plt.legend()
    
```

```python
for i in [1,2,5,10,50,100,500]:
    get_plot(k, theta,i)
```
